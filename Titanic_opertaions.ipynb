```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("train.csv")

df.head()
```

```python
survival_rates = df.pivot_table(values="Survived", index="Pclass", columns="Sex", aggfunc="mean")

plt.figure(figsize=(8,6))
sns.barplot(x="Pclass", y="Survived", hue="Sex", data=df)
plt.title("Survival Rate by Gender & Class")
plt.ylabel("Survival Rate")
plt.show()

print(survival_rates)
```

```python
df["FamilySize"] = df["SibSp"] + df["Parch"] + 1

df["FamilyCategory"] = pd.cut(df["FamilySize"], bins=[0,1,4,7], labels=["Alone", "Small Family", "Large Family"])

plt.figure(figsize=(8,6))
sns.barplot(x="FamilyCategory", y="Survived", data=df)
plt.title("Survival Rate by Family Size")
plt.ylabel("Survival Rate")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x="Embarked", y="Survived", data=df)
plt.title("Survival Rate by Embarkation Port")
plt.ylabel("Survival Rate")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="Survived", y="Fare", data=df)
plt.title("Fare Paid vs. Survival")
plt.ylabel("Fare Amount")
plt.show()
```
